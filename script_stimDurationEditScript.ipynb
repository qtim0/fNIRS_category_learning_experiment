```python
%reset
from snirf import Snirf
import numpy as np
import pandas as pd
import os
```

```python
#this gets filenames, you'll have to store em in a list or something but we're almost there!!!
filenames = []
for file in os.listdir("C:/Users/user00/My Drive/category_fNIRS/exp3_analysis/snirfs_edit"):
    if file.endswith(".snirf"):
        filenames.append(file)

#print(filenames)
print(len(filenames))
```

Now we have to write a loop in which each .snirf is read in, then the stim array is pulled out for all three conditions catg, ctrl, rest. They are modified to hold the correct values. Finally, the stim array is replaced with the correct stims. It is written to the .snirf.

By the way, Homer3 will make the files 'invalid' unless they are saved at least once through pysnirf2 no fucking idea why it is just the case.

```python
#Declare proper duration vectors
catgdur = [48.0, 48.0, 48.0, 48.0, 48.0]
ctrldur = [48.0, 48.0, 48.0, 48.0, 48.0]
restdur = [20.0, 12.0, 20.0, 14.0, 18.0, 16.0, 16.0, 18.0, 14.0, 20.0, 12.0]

#get path
#tpath = os.getcwd()
#print(tpath)
tpath = "c:\\Users\\user00\\My Drive\\category_fNIRS\\exp3_analysis\\snirfs_edit\\"
```

```python
#loop through and modify each file
for i in filenames:

    lpath = tpath + i  #create path to each file
    print(lpath)

    dataf = Snirf(lpath, 'r+')   #Read in the snirf file in edit mode

    catg = dataf.nirs[0].stim[0].data  #get incorrect stim array
    ctrl = dataf.nirs[0].stim[1].data
    rest = dataf.nirs[0].stim[3].data

    catg[:,1] = catgdur     #assign correct duration values to col2
    ctrl[:,1] = ctrldur
    rest[:,1] = restdur

    dataf.nirs[0].stim[0].data = catg    #write correct stim array
    dataf.nirs[0].stim[1].data = ctrl
    dataf.nirs[0].stim[3].data = rest

    print(i)  #print diagnostics
    dataf.validate().display(severity = 3)
    
    dataf.save()  #save to file
    dataf.close()

    del dataf #clear variables for next iteration
    del lpath
    del catg
    del ctrl
    del rest
```
